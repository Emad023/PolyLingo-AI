```python
async def process_text_to_speech(file_type: str, file_path: str, target_lang: str, gender: str) -> str:
    text = extract_text_(file_type, file_path)
    if not text or not text.strip():
        raise ValueError("Extracted text is empty.")

    print("\nOriginal Text:\n", text)

    translated_text = translate_text(text, dest_language=target_lang)
    print("\nTranslated Text:\n", translated_text)

    print("\nTranslated Audio")
    await speak(translated_text, lang=target_lang, gender=gender)
```
