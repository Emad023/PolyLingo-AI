```python
async def process_audio(audio_file: str, target_lang: str, gender: str):
    audio_language = detect_language(audio_file)
    
    if audio_language == "en":
        print("\nTranscript:-")
        transcript = transcribe_english(audio_file)
    else:
        print("\nTranscribing:-")
        transcript = transcribe_with_detected_language(audio_file, audio_language)

    print(transcript)

    translated_text = translate_text(transcript, dest_language=target_lang)

    print("\nTranslated Text:-")
    print(translated_text)

    print("\nTranslated Audio\n")
    await speak(translated_text, lang=target_lang, gender=gender)

    return {
        "Detected_language": audio_language,
        "Transcript": transcript,
        "Translated_text": translated_text
    }
```
