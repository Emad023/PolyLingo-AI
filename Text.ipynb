```python
async def translate_text_and_speak(text: str, target_lang: str, gender: str):
    print("Original Text:\n", text)

    detected_lang = detect_language(text)
    print(f"\nDetected Language: {detected_lang}")

    translated_text = translate_text(text, dest_language=target_lang)
    print(f"\nTranslated Text ({target_lang}):\n", translated_text)

    print("\nSpeaking Translated Text...")
    await speak(translated_text, lang=target_lang, gender=gender)
```
