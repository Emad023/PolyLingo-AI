```python
def extract_text_(file_type: str, file_path: str) -> str:
    extracted_text = ""

    if file_type == "text":
        extracted_text = extract_text_from_txt(file_path)
    elif file_type == "pdf":
        extracted_text = extract_text_from_pdf(file_path)
    elif file_type == "image":
        try:
            import pytesseract
            from PIL import Image
        except ImportError:
            raise ImportError("pytesseract and Pillow must be installed for image text extraction.")
        extracted_text = extract_text_from_image(file_path)
    else:
        raise ValueError("Unsupported file type. Use 'txt', 'pdf' or 'image'.")

    if not extracted_text:
        raise ValueError("No text extracted from the file.")

    source_lang = detect_language(extracted_text)
    print(f"Detected source language: {source_lang}")

    return extracted_text
```
